```python
import csv

filename = "data203_新北市公共自行車即時資訊.csv"
available = [] #欄位：sbi 代表站點目前可提供的ubike
total = [] #欄位：'tot' 代表站點總共提供多少ubike

with open(filename, 'r', encoding="utf-8") as fp:
    data = csv.reader(fp)
    data = list(data)
    data.pop(0) #去掉欄位

    for row in data:
        total.append(int(row[2]))
        available.append(int(row[3]))
print(f"新北市UBIKE總量: {sum(total)}")
print(f"新北市UBIKE目前可使用量: {sum(available)}")
print(f"新北市UBIKE目前已被租借: {sum(total)-sum(available)}")
usage = ((sum(total)-sum(available))/sum(total))*100
print(f"新北市UBIKe使用率: {usage:.2f}%")
```

```python
import csv

filename = "data203_新北市公共自行車即時資訊.csv"
usable   = [] #儲存每筆資料，需要的欄位

with open(filename, 'r', encoding="utf-8") as fp:
    data = csv.reader(fp)
    data = list(data)
    data.pop(0) #去掉欄位
    
    for row in data:
        mytup = (row[0].strip(),row[1].strip(),eval(row[2]),int(row[3]),row[4].strip(),int(row[-1]))
        usable.append(mytup)
            

    
# 哪一站有最多ubike可以借

usable.sort(key = lambda x:x[3])
print(f"剩餘最多ubike可借: {usable[-1][3]}, 區域: {usable[-1][4]}, 站點名稱: {usable[-1][1]}")


# 哪一站目前沒有ubike可以借

unable = []
for row in usable:
    if row[3]==0:
        unable.append(row)
print(f"Ubike剩餘數量為0 共有{len(unable)}個站點")  

for row in unable:
    print(f"站點區域: {row[4]:<6} 站點名稱: {row[1]:<15} 總量: {row[2]:<6} 可租借量: {row[3]:<5}")
print("="*60)

#哪一些站不能使用
unable_sta = []
for row in usable:
    if row[-1]!=1:
        unable_sta.append(row)
print(f"目前不提供服務的站點: {len(unable_sta)}")
```

```python
#4. 新北市每一區分別有多少ubike站點 list tuple set dict
# mydict = {區域名稱1:{站點名稱:[總量,可租借量],站點名稱2:[總量,可租借量],站點名稱3:[總量,可租借量]}
#           區域名稱2:{站點名稱:[總量,可租借量],站點名稱2:[總量,可租借量],站點名稱3:[總量,可租借量]}

# 區域,多少站點,可租借ubike總量,站點名稱(Group by)
# row[4]: 區域名稱


mydict = {}
for row in usable:
    if row[4] not in mydict: #區域是否為第一次遇見判斷
        mydict[row[4]] = {row[1]:[row[2],row[3]]}
    else:
        mydict[row[4]][row[1]] = [row[2],row[3]]
        #mydict[row[4]]=>取得區域為key的value{}
        #mydict[row[4]][row[1]]:因為又是一個字典{},站點名稱為key        

filename = "district_start.csv"
        
for dis in mydict:
    print("區域名稱:",dis, "站點數量",len(mydict[dis]))
    station = []
    bike_total = 0
    for s in mydict[dis]:
        station.append(s)
        bike_total += mydict[dis][s][0]
        
    with open(filename, "a",encoding="big5") as fp:
        w = csv.writer(fp,delimiter=",")
        w.writerow([dis,len(mydict[dis]),bike_total])
        
    print("區域名稱:",dis, "站點數量",len(mydict[dis]), "區域總量:",bike_total)
    print("站點名稱")
    print(*station)
    print("="*60)
```
